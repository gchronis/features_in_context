```python
import pandas as pd
df = pd.read_csv('../results/lappin_clusters_nouns_only.csv')
df
```

```python
samples = []

for noun in df.word.unique():
    sentences = df[df['word'] == noun]
    length = len(sentences)
    #print(noun)
    #print(length)
    for cluster_id in range (0,5):
        cluster_sentences = sentences[sentences['cluster_id'] == cluster_id]
        #print(cluster_sentences)
        sample = cluster_sentences.sample(random_state=1)
        samples.append(sample)

samples = pd.concat(samples)
samples
```

```python
for noun in df.word.unique():
    this_word = samples[samples['word'] == noun]
    print(noun)
    print()
    for index, row in this_word.iterrows():
        print(row.sentence)
        
        for model in models.values():
            features = model.predict_top_n_features_in_context(noun, row.sentence, 10, bert=bert)
            print(features)
        print()
        print()
```
