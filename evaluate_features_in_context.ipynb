```python
import csv
from pandas import DataFrame
import pandas as pd


sents = []
with open('data/duck.txt') as csvfile:
    reader = csv.DictReader(csvfile, fieldnames=["sentence", "sense_id"], delimiter='\t')
    for row in reader:
        sents.append(row)
        print(row['sentence'], row['sense_id'])
df = DataFrame.from_dict(sents)

print(df)
```

```python
pd.set_option('display.max_rows', 1000)

feats = []

for ind in df.index:
    predicted_features = mpro_model.predict_top_n_features_in_context('duck', df['sentence'][ind], 7, bert=bert)
    feats.append(predicted_features)


    
df['features'] = feats

df
```
